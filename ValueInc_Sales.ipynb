```python
NumberOfItemsPurchased = 6
CostPerItem = 11.73
SellingPricePerItem = 21.11
ProfitPerItem = SellingPricePerItem - CostPerItem
print ("ProfitPerItem = ",ProfitPerItem)
ProfitOnEachTranssaction = ProfitPerItem * NumberOfItemsPurchased
print ("ProfitOnEachTranssaction = ",ProfitOnEachTranssaction)
TotalCost = CostPerItem * NumberOfItemsPurchased
print ("TotalCost = ",TotalCost)
TotalProfit = SellingPricePerItem * NumberOfItemsPurchased
print ("TotalProfit = ",TotalProfit)
```

```python
df['CostPerTransaction'] = df['CostPerItem'] * df ['NumberOfItemsPurchased']
```

```python
df.head()
```

```python
df['SalesPerTransaction'] = df['SellingPricePerItem'] * df ['NumberOfItemsPurchased']
```

```python
df.head()
```

```python
df['ProfitPerTransaction'] = (df['SellingPricePerItem'] - df ['CostPerItem']) * df ['NumberOfItemsPurchased']
```

```python
df.head()
```

```python
df['Markup'] = df['ProfitPerTransaction'] / df['CostPerTransaction']
```

```python
df.head()
```

```python
df['Markup'] = round (df['Markup'] , 2)
df.head()
```

```python
df['Date'] = df['Day'].astype(str) + '-' + df['Month'].astype(str) + '-' + df['Year'].astype(str)
```

```python
df.head()
```

```python
df.iloc[2, 2]
```

```python
split_col = df['ClientKeywords'].str.split(',' , expand = True)
split_col
```

```python
df['ClientAge'] = split_col[0]
df['ClientType'] = split_col[1]
df['LengthofContract'] = split_col[2]
df.head()
```

```python
df['ClientAge'] = df['ClientAge'].str.replace('[', '', regex=False)
```

```python
df['LengthofContract'] = df['LengthofContract'].str.replace(']', '', regex=False)
df.head()
```

```python
df['ItemDescription'] = df['ItemDescription'].str.lower()
```

```python
season= pd.read_csv('value_inc_seasons.csv' ,  sep = ';')
season.head()
```

```python
df = pd.merge(df , season , on = 'Month' )
```

```python
df.head()
```

```python
df = df.drop(['Year','Month','Day', 'ClientKeywords'] , axis = 1)
df.head()
```
