```python
df = df.drop(['engagement_comment_plugin_count'], axis = 1)
df.head()
```

```python
def keywordflag(keyword):
    length = len(df)
    keyword_flag=[]
    for i in range(0,length):
        heading = df['title'][i]
        try:
            if keyword in heading:
                flag=1
            else:
                flag=0
        except:
            flag=0
        keyword_flag.append(flag)
    return keyword_flag
k = keywordflag('murder')
```

```python
df['keyword_flag'] = pd.Series(k)
df.head()
```

```python
sent_int = SentimentIntensityAnalyzer()
```

```python
text = df['title'][0]
text
```

```python
sent = sent_int.polarity_scores(text)
sent
```

```python
title_neg_sentiment = []
title_pos_sentiment = []
title_neu_sentiment = []
length = len(df)
for i in range (0,length):
    try:
        text = df['title'][i]
        sent = sent_int.polarity_scores(text)
        neg = sent['neg']
        pos = sent['pos']
        neu = sent['neu']
    except:
        neg = 0
        pos = 0
        neu = 0
    title_neg_sentiment.append(neg)
    title_pos_sentiment.append(pos)
    title_neu_sentiment.append(neu)
```

```python
title_neg_sentiment = pd.Series(title_neg_sentiment)
title_pos_sentiment = pd.Series(title_pos_sentiment)
title_neu_sentiment = pd.Series(title_neu_sentiment)
```

```python
df['title_neg_sentiment'] = title_neg_sentiment
df['title_pos_sentiment'] = title_pos_sentiment
df['title_neu_sentiment'] = title_neu_sentiment
```
