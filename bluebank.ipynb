```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
js = open ('loan_data_json.json')
```

```python
data = json.load(js)
```

```python
df= pd.DataFrame(data)
df.head()
```

```python
df['purpose'].unique()
```

```python
df.describe()
```

```python
df['annualincome'] = np.exp(df['log.annual.inc'])
df.head()
```

```python
def get_fico_status(fico):
    if isinstance(fico, str):
        return 'Unknown'
    fico = int(fico)
    if fico >= 700:
        return 'Excellent'
    elif 700 > fico >= 660:
        return 'Good'
    elif 660 > fico >= 601:
        return 'Fair'
    elif 601 > fico >= 400:
        return 'Poor'
    elif 400 > fico >= 300:
        return 'Very Poor'
    else:
        return 'Very Poor'

df['fico.category'] = df['fico'].apply(get_fico_status)
df.head()
```

```python
df.loc[df['int.rate'] > 0.12 , 'int.rate.type'] = 'High'
df.loc[df['int.rate'] <= 0.12 , 'int.rate.type'] = 'Low'
df.head()
```

```python
df['fico.category'].value_counts()
```

```python
ficostatus_counts = df['fico.category'].value_counts()
plt.figure(figsize=(10, 6)) 
sns.barplot(x=ficostatus_counts.index, y=ficostatus_counts.values, palette="viridis")
```

```python
purpose_counts = df['purpose'].value_counts()
plt.figure(figsize=(15, 7)) 
sns.barplot(x=purpose_counts.index, y=purpose_counts.values, palette="viridis")
```

```python
df.to_csv('loan_cleaned.csv' , index = True)
```
